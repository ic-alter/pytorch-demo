## 使用大模型API进行推理

在这个项目中可以申领有一定免费额度的API key，可使用gpt-3.5-turbo等模型：https://github.com/chatanywhere/GPT_API_free

```python
import requests

API_KEY = "想要体验此API请联系助教获取API KEY"
BASE_URL = "https://mapi.fduer.com/api/v1"
model = "deepseek-r1-671b"
```

```python
def generate_text(prompt):
    url = f"{BASE_URL}/chat/completions"
    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }
    data = {
        "model": model,   # 根据服务器支持的模型替换
        "messages": [
            {"role": "user", "content": prompt}
        ],
        "max_tokens": 100,
        "temperature": 0.7
    }

    resp = requests.post(url, headers=headers, json=data)
    if resp.status_code == 200:
        result = resp.json()
        return result["choices"][0]["message"]["content"]
    else:
        raise Exception(f"Error {resp.status_code}: {resp.text}")

prompt = "请用一句话介绍一下PyTorch是什么。"
output = generate_text(prompt)
print("=== 模型输出 ===")
print(output)
```
